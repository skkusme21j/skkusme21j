## T-test or Anova test

### 직무열의

#### - 성별 간 t test

```python
M = data.loc[data['gender']==1, ['Job_engagement']]
F = data.loc[data['gender']==2, ['Job_engagement']]
```

```python
# 정규성 만족
from scipy.stats import shapiro
normal1 = shapiro(M.values[:, 0])
normal2 = shapiro(F.values[:, 0])
print(normal1, normal2)
# 정규성을 만족하지 X 대표본이기에 크게 고려 X
```

```python
M.values[:, 0]
```

```python
#levene test로 등분산성을 검정  # 정규성 관계 X
from scipy.stats import levene, ttest_ind
print(levene(M.values[:, 0],F.values[:, 0]))

#등분산성을 bartlett test로 할 수도 있음 바틀렛 # 정규성을 만족할 때만
from scipy.stats import bartlett
print(bartlett(M.values[:, 0],F.values[:, 0]))
```

```python
import scipy.stats

ttest_ind(M.values[:, 0],F.values[:, 0], equal_var=False)
# 성별 간 직무열의 평균에 유의미한 차이 존재
```

#### - 연령별 ANOVA test

```python
A_20 = data.loc[data['AGE2']==1, ['Job_engagement']]
A_30 = data.loc[data['AGE2']==2, ['Job_engagement']]
A_40 = data.loc[data['AGE2']==3, ['Job_engagement']]
A_50 = data.loc[data['AGE2']==4, ['Job_engagement']]
```

```python
import scipy.stats as stats

    
# 정규성 만족
normal1 = shapiro(A_20.values[:, 0])
normal2 = shapiro(A_30.values[:, 0])
normal3 = shapiro(A_40.values[:, 0])
normal4 = shapiro(A_50.values[:, 0])
print(normal1, normal2, normal3, normal4)
# 정규성을 만족하지 X 대표본이기에 크게 고려 X
```

```python
#levene test로 등분산성을 검정  # 정규성 관계 X
from scipy.stats import levene
print(levene(A_20.values[:, 0], A_30.values[:, 0], A_40.values[:, 0], A_50.values[:, 0]))

#등분산성을 bartlett test로 할 수도 있음 바틀렛 # 정규성을 만족할 때만
from scipy.stats import bartlett
print(bartlett(A_20.values[:, 0], A_30.values[:, 0], A_40.values[:, 0], A_50.values[:, 0]))

# 등분산 X
```

```python
## ANOVA 검정
F, p = stats.f_oneway(A_20.values[:, 0], A_30.values[:, 0], A_40.values[:, 0], A_50.values[:, 0])
print( 'F-Ratio: {}'.format(F)
    , 'p-value:{}'.format(p)
     , sep = '\n')
```

```python
from statsmodels.sandbox.stats.multicomp import MultiComparison
from scipy.stats import ttest_ind
 
comp = MultiComparison(data['Job_engagement'], data['AGE2'])
ret = comp.allpairtest(scipy.stats.ttest_ind, method='bonf')
print(ret[0])
```

```python
import pandas as pd 
import numpy as np
import statsmodels.api as sm


# crim, rm, lstat을 통한 다중 선형회귀분석
x_data = test_df_car[["DJ","IJ","PJ"]] #변수 여러개
Health = test_df_car[["HE"]]

# for b0, 상수항 추가
x_data1 = sm.add_constant(x_data, has_constant = "add")

# OLS 검정
multi_model = sm.OLS(Health, x_data1)
fitted_multi_model = multi_model.fit()
fitted_multi_model.summary()
```
